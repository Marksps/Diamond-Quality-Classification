```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, make_scorer
```

```python
def percent(a, b) :
    result = 100 - abs(float(((b - a) * 100) / a))
    return result
```

```python
def getDifference(b1, b2):
	r = (b2 - b1) % 360.0
	
	if r >= 180.0:
		r -= 360.0
	return r
```

```python
def find_area_half(path):
  df = pd.DataFrame(columns = ["area1","area2"])

  for j in range(1,9):
    all_area = []
    
    for i in range(1,3):
      area = []
      link = path + str(j)+'-'+str(i)+'.png'
      img = cv.imread(link,0)
      hist = cv.calcHist([img],[0],None,[256],[0,256])

      # Area occupied by black region
      black_area = np.true_divide(hist[0],np.prod(img.shape))[0]*100

      # extract no black parts
      thresh = cv.threshold(img,200,255,cv.THRESH_BINARY)[1]
      kernel = np.ones((3,3),np.uint8)

      # fill in the small white spots
      opening = cv.morphologyEx(thresh, cv.MORPH_OPEN, kernel)

      # extract the contours
      contours = cv.findContours(opening, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)[0]

      blank_image = np.zeros((img.shape),np.uint8)
      image_area = np.prod(img.shape)

      # iterate through the contours detected from right top corner
      for i,c in enumerate(contours[::-1]):

          # turn blank_image black
          blank_image *= 0

          # draw filled contour
          cv.drawContours(blank_image, [c], 0, (255), thickness=cv.FILLED)

          contour_area = cv.contourArea(c)
          area.append(contour_area)
          # percentage of area contour
          contour_area_pc = np.true_divide(int(contour_area),image_area)*100 if int(contour_area) > 1  else 0 

        
      area.sort(reverse=True)
      try:
        all_area.append((area[0] + area[1]))
      except:
        try:
          all_area.append((area[0]))
        except:
          all_area.append((None))

    to_append = all_area
    a_series = pd.Series(to_append, index = df.columns)
    df = df.append(a_series, ignore_index=True)

  acc = []
  for i in range(len(df['area1'])):
    percent_heart = percent(df['area1'][i], df['area2'][i])
    acc.append(percent_heart)
    #print('Heart',(i+1),'Symmetry Score : %.2f'%percent_heart,'%')
  print('------------------------------------------')
  print('Average of Symmetry Score : %.2f' %mean(acc),'%')
```

```python
def angular_distance(angle):
  ang_dist = []
  ang_sc = []
  for i in range(0,7):
    sc = getDifference(angle[i], angle[i+1])
    ang_dist.append(abs(sc))
    #print('Angular Distance between',i+1,'and',i+2,': %.2f'%abs(sc))
  sc81 = getDifference(angle[7], angle[0])
  ang_dist.append(abs(sc81))
  #print('Angular Distance between',8,'and',1,': %.2f'%abs(sc81))
  #print('------------------------------------------')
  for i in range(len(ang_dist)):
      for j in range(i + 1, len(ang_dist)):
        sc = percent(ang_dist[i], ang_dist[j])
        ang_sc.append(sc)
        #print('Angular Distance Score between Heart',i+1,'and',j+1,'is %.2f'%sc,'%')
  print('------------------------------------------')
  return print('Average Angular Distance Score : %.2f'%mean(ang_sc),'%')
```

```python
def area_score(area):
  area_score = []

  for i in range(len(area)):
      for j in range(i + 1, len(area)):
        sc = percent(area[i], area[j])
        area_score.append(sc)
        #print('Area Score between Heart',i+1,'and',j+1,'is %.2f'%sc,'%')
        
  print('------------------------------------------')
  print('Average Area Score : %.2f'%mean(area_score),'%')
  print('------------------------------------------')
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbpipeline
from sklearn.pipeline import Pipeline
from sklearn.datasets import make_classification, load_breast_cancer

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=11)

smote = SMOTE(random_state = 1)
X_train, y_train = smote.fit_resample(X_train, y_train)

pipeline = imbpipeline(steps = [['scaler', MinMaxScaler()],
                                ['classifier', LogisticRegression(random_state=11, max_iter=1000)]])

stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=11)
    
param_grid = {'classifier__C':[0.001, 0.01, 0.1, 1, 10, 100, 1000]}
grid_search = GridSearchCV(estimator=pipeline,
                           param_grid=param_grid,
                           cv=stratified_kfold,
                           n_jobs=-1)

grid_search.fit(X_train, y_train)
cv_score = grid_search.best_score_
test_score = grid_search.score(X_test, y_test)

print('Cross-validation score: %.2f' % cv_score)
print('Test score: %.2f' % test_score)
```
