```python
def percent(a, b) :
    result = 100 - abs(float(((b - a) * 100) / a))
    return result
```

```python
from statistics import mean
```

```python
acc = []
for i in range(len(df['area1'])):
  percent_heart = percent(df['area1'][i], df['area2'][i])
  acc.append(percent_heart)
  print('Heart',(i+1),'Symmetry Score : %.2f'%percent_heart,'%')
print('------------------------')
print('Average of Symmetry Score : %.2f' %mean(acc),'%')
```

## Angular Distance

```python
angle = [67.06789956,	22.40212959,	-22.40212959,	-67.66455743,	-112.8736652,	-158.3495672,	157.0678996,	111.8791513]
```

```python

def getDifference(b1, b2):
	r = (b2 - b1) % 360.0
	
	if r >= 180.0:
		r -= 360.0
	return r
```

```python
getDifference(67.06789956,22.40212959)
```

```python
def angular_distance(angle):
  ang_dist = []
  ang_sc = []
  for i in range(0,7):
    sc = getDifference(angle[i], angle[i+1])
    ang_dist.append(abs(sc))
    print('Angular Distance between',i+1,'and',i+2,': %.2f'%abs(sc))
  sc81 = getDifference(angle[7], angle[0])
  ang_dist.append(abs(sc81))
  print('Angular Distance between',8,'and',1,': %.2f'%abs(sc81))
  print('------------------------------------------')
  for i in range(len(ang_dist)):
      for j in range(i + 1, len(ang_dist)):
        sc = percent(ang_dist[i], ang_dist[j])
        ang_sc.append(sc)
        print('Angular Distance Score between Heart',i+1,'and',j+1,'is %.2f'%sc,'%')
  print('------------------------------------------')
  return print('Average Angular Distance Score : %.2f'%mean(ang_sc),'%')
```

```python
angular_distance(angle)
```

```python
def area_score(area):
  area_score = []
  for i in range(0,7):
    print('Heart',(i+1),'Area : %.2f' %area[i])
  print('------------------------------------------')

  for i in range(len(area)):
      for j in range(i + 1, len(area)):
        sc = percent(area[i], area[j])
        area_score.append(sc)
        print('Area Score between Heart',i+1,'and',j+1,'is %.2f'%sc,'%')
        
  print('------------------------------------------')

  return print('Average Area Score : %.2f'%mean(area_score),'%')
```
